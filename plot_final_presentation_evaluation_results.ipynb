```python
#!/usr/bin/env python3
#Plot pre-processing misses
import csv
import re
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount("/content/drive", force_remount=True)

source_data = '/content/drive/MyDrive/ColabNotebooks/FinalProject_AI/Source_Data/'
source_data_output = '/content/drive/MyDrive/ColabNotebooks/FinalProject_AI/Source_Data_Output/'
source_code = '/content/drive/MyDrive/ColabNotebooks/FinalProject_AI/Source_Code/'
admin = '/content/drive/MyDrive/ColabNotebooks/FinalProject_AI/Admin/'
backups = '/content/drive/MyDrive/ColabNotebooks/FinalProject_AI/Backups/'

misses_by_count = source_data +'plot_exclusions_csv.csv'
misses_by_percentage = source_data + 'preprocessing_misses_csv.csv'
tokens_vs_accuracy = source_data +'TokenSize_vs_Accuracy.csv'

```

```python
results_df = pd.read_csv(misses_by_percentage)

# Plot the results
plt.figure(figsize=(10, 6))
bars = plt.bar(results_df['Token Size'], results_df['Miss Percentage'], color='blue')
plt.xlabel('Token Sizes')
plt.ylabel('Miss Percentage (%)')
plt.title('Pre-Processing Miss Percentage')


# Annotate the bars with the miss percentage text
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}%', ha='center', va='bottom')

# Set the x-axis tick labels to the token sizes
#show all 10 xticks
plt.xticks(range(1, 6), results_df['Token Size'])
plt.tight_layout()
plt.show()

```

```python
results_df = pd.read_csv(misses_by_count)
 # Plot the results
plt.figure(figsize=(10, 6))
bars = plt.bar(results_df['Token Size'], results_df['Count'], color='blue')
plt.xlabel('Token Sizes')
plt.ylabel('Excluded Count')
plt.title('Pre-Processing Exclude Count')


# Annotate the bars with the miss percentage text
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:}', ha='center', va='bottom')

# Set the x-axis tick labels to the token sizes
#show all 10 xticks
plt.xticks(range(1, 6), results_df['Token Size'])
plt.tight_layout()
plt.show()
```

```python
results_df = pd.read_csv(tokens_vs_accuracy)

# Plot the results
plt.figure(figsize=(10, 6))
bars = plt.bar(results_df['Token Size'], results_df['Prediction Accuracy'], color='blue')
plt.xlabel('Token Sizes')
plt.ylabel('Accuracy(%)')
plt.title('Tokens vs Prediction Accuracy')


# Annotate the bars with the miss percentage text
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}%', ha='center', va='bottom')

# Set the x-axis tick labels to the token sizes
#show all 10 xticks
plt.xticks(range(1, 6), results_df['Token Size'])
plt.tight_layout()
plt.show()
```
